## Scrape NBA player Tweets

```python
#source : https://stackoverflow.com/questions/67939869/python-how-to-iterate-through-a-list-of-twitter-users-using-snscrapen ;/
```

```python
import snscrape.modules
# year 2022
# list to iterate over 
users_name = player_2022
# create list to append tweets to 
tweet_list1 = []
mediaurl=[]

## Adjust date from each year needed 
for n, k in enumerate(users_name):
    #print(n)
    #print(k)
    #for i,tweet in enumerate(sntwitter.TwitterSearchScraper('from:{}'.format(users_name[n]) + ' since:2022-01-01 until:2022-6-31' ).get_items()): 
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper('from:{}'.format(users_name[n]) + ' since:2022-01-01 until:2022-12-31').get_items()):
        if i>1:
            break
        tweet_list1.append([tweet.date, tweet.id, tweet.content,\
                         tweet.user.username, tweet.user.followersCount,\
                        tweet.retweetCount, tweet.likeCount,\
                        tweet.retweetedTweet, tweet.quotedTweet,\
                       tweet.inReplyToUser, tweet.mentionedUsers, tweet.hashtags, tweet.url,tweet.media, tweet.user.rawDescription ])
        if tweet.media:
            for medium in tweet.media:
                if isinstance(medium, snscrape.modules.twitter.Photo):
                    mediaurl.append([medium.fullUrl, tweet.id, tweet.content])
                elif isinstance(medium, snscrape.modules.twitter.Video):
                    for v in medium.variants:
                        mediaurl.append([v.url.replace("?tag=13", "").replace("?tag=10", ""), tweet.id, tweet.content])
```

```python
#tweet_list1
```

```python
tweets_df1 = pd.DataFrame(tweet_list1, columns = [['date','id','tweet','username',\
                                                 'follow_count','retweet_count','like_count',\
                                                 'retweeted_tweet','qutoted_tweet',\
                                                 'repled_user','mentioned_user','hashtag',\
                                                  'url','media','media_url_photo']])
tweets_df1.shape
```
